```

import requests
import csv
from bs4 import BeautifulSoup as bs
import urllib 
import os
import requests 
import pandas as pd
```

```
url= 'https://www.goodreads.com/review/list/10783335-t-s?shelf=read'  
```

```
#Page 1, 30 books
response=requests.get(url)
#Store the contents of the website under doc

soup = bs(response.text, 'html.parser')
print(soup.prettify())
```

```
book_table = soup.find('table', class_ = 'table stacked')
book_table
```

```
type(book_table)
```

```
headers= [header.text for header in book_table.find_all('th')]
headers
```

```

rows = []
for row in book_table.find_all('tr'):
     rows.append([val.text.encode('utf8') for val in row.find_all('td')])
      
headers = [i.replace("\n", "") for i in headers]
# rows = [i.replace("\n", "") for i in rows]
```

```

df1 = pd.DataFrame(rows, columns = headers)
df1.head()
```
